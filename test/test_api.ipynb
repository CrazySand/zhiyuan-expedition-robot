## Robot API

```python
client = httpx.AsyncClient(timeout=60, base_url="http://127.0.0.1:59888")
```

```python
# 启动人脸识别（POST = 启动）
response = await client.post("/api/face-recognition")
response.json()
```

```python
# 停止人脸识别（DELETE = 停止）
response = await client.delete("/api/face-recognition")
response.json()
```

```python
# 查询人脸识别进程状态（GET = 状态）
response = await client.get("/api/face-recognition")
response.json()
```
