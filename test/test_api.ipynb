### 通用接口（POST /api/common）

所有可编排接口收敛为单一入口，请求体：`{ "action": "动作名", "params": { ... } }`。使用上面同一 `client`（base_url 为 PC 端 8000）。

```python
# 获取支持的 action 列表
response = await client.get("/api/common/actions")
response.json()
```

```python
# tts.play 发起 TTS 播报
response = await client.post("/api/common", json={"action": "tts.play", "params": {"text": "你好，这是通用接口测试"}})
response.json()
```

```python
# tts.stop 打断 TTS 播报
response = await client.post("/api/common", json={"action": "tts.stop", "params": {}})
response.json()
```

```python
# tts.status 查询 TTS 播报状态（trace_id 可用上面 tts.play 返回的）
response = await client.post("/api/common", json={"action": "tts.status", "params": {"trace_id": "hafhjkqwjwefk-6bC8EcYq6GaURi7KJUMM2O"}})
response.json()
```

```python
# tts.volume.get 获取音量
response = await client.post("/api/common", json={"action": "tts.volume.get", "params": {}})
response.json()
```

```python
# tts.volume.set 设置音量（0~70）
response = await client.post("/api/common", json={"action": "tts.volume.set", "params": {"audio_volume": 50}})
response.json()
```

```python
# agent.properties.set 设置交互模式并重启 agent（only_voice | voice_face | normal）
response = await client.post("/api/common", json={"action": "agent.properties.set", "params": {"mode": "normal"}})
response.json()
```

```python
# agent.properties.get 查询交互模式
response = await client.post("/api/common", json={"action": "agent.properties.get", "params": {}})
response.json()
```

## Robot API

```python
client = httpx.AsyncClient(timeout=60, base_url="http://127.0.0.1:59888")
```

```python
# 启动人脸识别（POST = 启动）
response = await client.post("/api/face-recognition")
response.json()
```

```python
# 停止人脸识别（DELETE = 停止）
response = await client.delete("/api/face-recognition")
response.json()
```

```python
# 查询人脸识别进程状态（GET = 状态）
response = await client.get("/api/face-recognition")
response.json()
```
