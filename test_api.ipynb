```python
import httpx
```

```python
headers = {
    "Authorization": "Bearer your-secret-key-here"
}
client = httpx.AsyncClient(timeout=60, headers=headers, base_url="http://127.0.0.1:8000")
# client = httpx.AsyncClient(timeout=10, headers=headers, base_url="http://39.108.111.105:8000")
```

### TTS

```python
response = await client.post("/api/play_tts", json={"text": "测试音频" * 20})
response.json()
```

```python
response = await client.post("/api/stop_tts")
response.json()
```

```python
resposne = await client.get("/api/get_audio_status", params={"trace_id": "hafhjkqwjwefk_6bC8EcYq6GaURi7KJUMM2O"})
resposne.json()
```

```python
resposne = await client.get("/api/get_audio_volume")
resposne.json()
```

```python
response = await client.post("/api/set_audio_volume", json={"audio_volume": 60})
response.json()
```
