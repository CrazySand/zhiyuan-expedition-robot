```python
import httpx
```

```python
headers = {
    "Authorization": "Bearer your-secret-key-here"
}
client = httpx.AsyncClient(timeout=60, headers=headers, base_url="http://127.0.0.1:8000")
# client = httpx.AsyncClient(timeout=10, headers=headers, base_url="http://39.108.111.105:8000")
```

### TTS Service

```python
response = await client.post("/api/play_tts", json={"text": "章晨与，你这支臭鸵鸟"})
response.json()
```

```python
response = await client.post("/api/stop_tts")
response.json()
```

```python
resposne = await client.get("/api/get_audio_status", params={"trace_id": "hafhjkqwjwefk_250udq0gD1Opu9xc3YU2CQr"})
resposne.json()
```

```python
resposne = await client.get("/api/get_audio_volume")
resposne.json()
```

```python
response = await client.post("/api/set_audio_volume", json={"audio_volume": 70})
response.json()
```
